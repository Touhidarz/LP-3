```python
import matplotlib.pyplot as plt
```

```python
curr_x = 2
delta_x = 1
precision = 0.000001
rate = 0.01
iter_counter = 1
max_iter = 10000
```

```python
def slope(x):
    return 2*(x+3)

def y_value(x):
    return (x+3)**2

x = []
y = []

# y.append(y_value(curr_x))
# x.append(curr_x)
```

```python
while (delta_x > precision and iter_counter < max_iter):
    prev_x = curr_x
    curr_x = prev_x - rate * slope(prev_x)
    y.append(y_value(curr_x))
    x.append(curr_x)
    delta_x = abs(curr_x - prev_x)

    print(f"Iteration :{iter_counter}")
    print(f"X : {curr_x}")
    iter_counter += 1

print(f"\nLocal minima : {curr_x}")
```
